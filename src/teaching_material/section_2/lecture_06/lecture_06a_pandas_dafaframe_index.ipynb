## 4. Select (Locate) based on index, or "labels", or "position".
Recommended [answer](https://stackoverflow.com/a/31593712)  
Please understand the difference of ```.loc``` and ```.iloc```

### Slice rows

````python
# first row
df.iloc[0]
````

````python
df.loc[0]  # same result different method. Be careful!
````

````python
# second row
df.iloc[1]
````

````python
# iloc gets rows (and/or columns) at integer locations. Works on position.
# Works on index, works with numbers.
# iloc is exclusive.
df.iloc[0:2]  # exclusive
````

````python
# .loc gets rows (and/or columns) with particular labels.
# .loc is label-based
#  (inclusive)
df.loc[0:2]
````

### Slice columns

````python
# locate by name. Single brackets. This is a Series.
df["Q2"]
````

````python
type(df["Q2"])  # One bracket => Series
````

````python
# Locate by name but use double brackets. This is a dataframe.
df[["Q2"]]
````

````python
type(df[["Q2"]])  # two brackets => dataframe
````

````python
df[["Q2", "Q5", "Q3"]]
````

### Slice rows and columns

````python
# all rows, and only columns with names'Q2':'Q3'
df.loc[:, 'Q2':'Q5']
````

````python
df.loc[0:10, 'Q2':'Q5']
````

````python
df.loc[0:4, 'Q3':'Q5']
````

````python
type(df.loc[:, 'Q2':'Q3'])
````

````python
# .loc is inclusiv, because it works with "names"
df.loc[0:3, 'Q2':'Q3']  # Inclusive, I write this many times, in case you missed it.
````

````python
df.loc[0:10, ['Q2', 'Q5']] 
````

````python
df.loc[8, 'Q2',] 
````

````python
# value of a single "object"
df.Q2[8]
````

## 5. Transpose. Sometimes it is needed.

````python
df.transpose().head(10)  # alternative syntax df.T
````

````python
# df.T
````

````python
transposed_df = df.T
transposed_df
````

````python
# this works
transposed_df.iloc[0:4]
````

````python
# this does not work!
transposed_df.loc["Q1": "Q2"]
````

````python
transposed_df.loc["Q1": "Q2", 1:3]
````

````python
transposed_df.index
````

````python
transposed_df.columns
````

````python
transposed_df.loc["Q1"]
````

````python
# this works
transposed_df.loc["Q1": "Q3"]
````

````python
# this works
transposed_df.loc[["Q1", "Q2", "Q3"]]
````
